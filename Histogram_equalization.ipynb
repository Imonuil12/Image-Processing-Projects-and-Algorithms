```python
import matplotlib.pyplot as plt
from skimage.exposure import match_histograms

source_image = plt.imread("R.jpeg")
reference_image = plt.imread("P.jpeg")

matched_image = match_histograms(reference_image, source_image, channel_axis=2)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 5), sharex=True, sharey=True)
ax1.axis('off')
ax2.axis('off')
ax3.axis('off')

ax1.imshow(reference_image)
ax1.set_title("Source")

ax2.imshow(source_image)
ax2.set_title("Reference")

ax3.imshow(matched_image)
ax3.set_title("Matched")

plt.tight_layout()
plt.show()
```
