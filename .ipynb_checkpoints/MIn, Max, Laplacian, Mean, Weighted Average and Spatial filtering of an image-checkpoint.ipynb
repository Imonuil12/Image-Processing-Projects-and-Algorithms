```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('Board.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Median Filtering
median_filtered = cv2.medianBlur(image, ksize=3)

# Spatial Filtering (using a simple kernel)
kernel = np.ones((3,3), np.float32) / 9
spatial_filtered = cv2.filter2D(image, -1, kernel)

# Mean/Average Filtering
mean_filtered = cv2.blur(image, (3,3))

# Weighted Average Filtering (using a Gaussian kernel)
weighted_avg_filtered = cv2.GaussianBlur(image, (3,3), 0)

# Min Filter
min_filtered = cv2.erode(image, kernel)

# Max Filter
max_filtered = cv2.dilate(image, kernel)

# Laplacian
laplacian = cv2.Laplacian(image, cv2.CV_64F)

# Display the results
plt.figure(figsize=(12, 8))

plt.subplot(3,3,1), plt.imshow(image), plt.title('Original')
plt.subplot(3,3,2), plt.imshow(median_filtered), plt.title('Median Filtered')
plt.subplot(3,3,3), plt.imshow(spatial_filtered), plt.title('Spatial Filtered')
plt.subplot(3,3,4), plt.imshow(mean_filtered), plt.title('Mean Filtered')
plt.subplot(3,3,5), plt.imshow(weighted_avg_filtered), plt.title('Weighted Avg Filtered')
plt.subplot(3,3,6), plt.imshow(min_filtered), plt.title('Min Filtered')
plt.subplot(3,3,7), plt.imshow(max_filtered), plt.title('Max Filtered')
plt.subplot(3,3,8), plt.imshow(laplacian), plt.title('Laplacian')

plt.tight_layout()
plt.show()
```
