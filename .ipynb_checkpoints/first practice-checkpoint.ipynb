```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

a1 = cv2.imread('log.jpg')
a = a1.astype(float) / 255.0
c = 3
f = c * np.log(1 + a)
f = (f * 255).astype(np.uint8)

# Create a figure with 1 row and 2 columns of subplots
fig, axes = plt.subplots(1, 2)

# Plot the first image on the first subplot
axes[0].imshow(cv2.cvtColor(a1, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')

# Plot the second image on the second subplot
axes[1].imshow(cv2.cvtColor(f, cv2.COLOR_BGR2RGB))
axes[1].set_title('Log Transformation Image')

plt.show()
```

```python
input_image = cv2.imread("MRI_Image.png")
gamma = 1.5
output_image = np.uint8(np.power(input_image / 255, gamma)*255)
cv2.imshow('Original Image', input_image)
cv2.imshow('Trnasformed Image', output_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
